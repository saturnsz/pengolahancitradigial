```python
import numpy as np 
```

```python
data = np.array([
    [10, 50, 100, 150],
    [200, 220, 30, 70],
    [120, 180, 20, 90],
    [160, 255, 40, 110]
], dtype=np.uint8)
```

```python
def t1(data, K):
    operasi = data.astype(np.float32) * K
    hasil = np.clip(operasi, 0, 255)
    return operasi, hasil 

def t2(data, C):
    operasi = data.astype(np.float32) + C
    hasil = np.clip(operasi, 0, 255)
    return operasi, hasil

def t3(data, K):
    operasi = data.astype(np.float32) * K
    hasil = np.clip(operasi, 0, 255)
    return operasi, hasil
```

```python
T1 = t1(data, 1.5)
print(f'Hasil Sebelum Clipping: \n{T1[0]} \nHasil Setelah Clipping: \n{T1[1]}')
```

```python
T2 = t2(T1[1], 50)
print(f'Hasil Sebelum Clipping: \n{T2[0]} \nHasil Setelah Clipping: \n{T2[1]}')
```

```python
T3 = t3(T2[1], 1.2)
print(f'Hasil Sebelum Clipping: \n{T3[0]} \nHasil Setelah Clipping: \n{T3[1]}')
```

```python
g = T3[1].astype(np.uint8)
print(f'Hasil Akhir setelah Peningkatan Kontras Ekstrem:\n{g}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.heatmap(data=T3[1].astype(int), cmap='gray', annot=True, fmt='d')
plt.title('Hasil Akhir (Output)')
plt.show()
```
