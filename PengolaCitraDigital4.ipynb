```python
import cv2
import matplotlib.pyplot as plt

# Baca gambar
img = cv2.imread('/content/drive/MyDrive/images.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Konversi BGR ke RGB agar warnanya benar

# Operasi transformasi
img_flip = cv2.flip(img, 1)

h, w = img.shape[:2]
center = (w/2, h/2)
angle = 35
scale = 1
m = cv2.getRotationMatrix2D(center, angle, scale)
img_rotate = cv2.warpAffine(img, m, (w, h))

img_resize = cv2.resize(img, (200, 200))
img_crop = img[100:300, 200:400]

# Tampilkan semua hasil
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].imshow(img_flip)
ax[0, 0].set_title('Image Flip')
ax[0, 0].axis('off')

ax[0, 1].imshow(img_rotate)
ax[0, 1].set_title('Image Rotate')
ax[0, 1].axis('off')

ax[1, 0].imshow(img_resize)
ax[1, 0].set_title('Image Resize')
ax[1, 0].axis('off')

ax[1, 1].imshow(img_crop)
ax[1, 1].set_title('Image Crop')
ax[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Membaca gambar dengan warna default OpenCV (BGR)
img = cv2.imread('/content/drive/MyDrive/images.jpeg')

# -----------------------------
# Operasi 1: Flip
# -----------------------------
img_flip = cv2.flip(img, 1)

# -----------------------------
# Operasi 2: Rotate
# -----------------------------
h, w = img.shape[:2]
center = (w/2, h/2)
angle = 35
scale = 1
m = cv2.getRotationMatrix2D(center, angle, scale)
img_rotate = cv2.warpAffine(img, m, (w, h))

# -----------------------------
# Operasi 3: Resize
# -----------------------------
img_resize = cv2.resize(img, (200, 200))

# -----------------------------
# Operasi 4: Crop
# -----------------------------
img_crop = img[100:300, 200:400]

# -----------------------------
# Menampilkan semua hasil (warna default BGR)
# -----------------------------
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].imshow(img_flip)
ax[0, 0].set_title('Image Flip (BGR)')
ax[0, 0].axis('off')

ax[0, 1].imshow(img_rotate)
ax[0, 1].set_title('Image Rotate (BGR)')
ax[0, 1].axis('off')

ax[1, 0].imshow(img_resize)
ax[1, 0].set_title('Image Resize (BGR)')
ax[1, 0].axis('off')

ax[1, 1].imshow(img_crop)
ax[1, 1].set_title('Image Crop (BGR)')
ax[1, 1].axis('off')

plt.tight_layout()
plt.show()
```

```python
imgcv = img_cv.copy()

imgcv_rgb = cv2.cvtColor(imgcv, cv2.COLOR_RGB2BGR)

fig, ax = plt.subplots(1,2, figsize=(8,16))
ax[0].imshow(imgcv)
ax[0].set_title('Image CV')
ax[1].imshow(imgcv_rgb)
ax[1].set_title('Image CV RGB')
```

```python
image = img_cv.copy()

imgcv_rgb = cv2.cvtColor(imgcv, cv2.COLOR_BGR2RGB)

img_hsv = cv2.cvtColor(imgcv, cv2.COLOR_BGR2HSV)

img_gray = cv2.cvtColor(imgcv, cv2.COLOR_BGR2GRAY)

_, binary = cv2.threshold(img_gray, 128, 255, cv2.THRESH_BINARY)

fig, ax = plt.subplots(2,2, figsize=(8,8))
ax[0,0].imshow(imgcv_rgb)
ax[0,0].set_title('Image RGB')
ax[0,1].imshow(img_hsv)
ax[0,1].set_title('Image HSV')
ax[1,0].imshow(img_gray, cmap='gray')
ax[1,0].set_title('Image Gray')
ax[1,1].imshow(binary, cmap='gray')
ax[1,1].set_title('Image Binary')

plt.tight_layout()
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

im = np.array(Image.open('/content/drive/MyDrive/images.jpeg'))

im_R = im.copy()
im_R[:, :, (1, 2)] = 0  # Set green and blue channels to 0

im_G = im.copy()
im_G[:, :, (0, 2)] = 0  # Set red and blue channels to 0

im_B = im.copy()
im_B[:, :, (0, 1)] = 0  # Set red and green channels to 0

fig, ax = plt.subplots(1, 3, figsize=(10, 10))

ax[0].imshow(im_R)
ax[0].set_title('Image R')
ax[0].axis('off')

ax[1].imshow(im_G)
ax[1].set_title('Image G')
ax[1].axis('off')

ax[2].imshow(im_B)
ax[2].set_title('Image B')
ax[2].axis('off')

plt.tight_layout()
plt.show()
```
