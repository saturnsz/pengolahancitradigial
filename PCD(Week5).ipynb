```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2 as cv
import numpy as np

img = cv2.imread('/content/drive/MyDrive/907_1.jpg', cv.IMREAD_COLOR_RGB)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(1,2,2)
plt.title('Histogram Intensitas')
plt.bar(range(256), hist.flatten(), color="green")
plt.xlabel('Tingkat Keabuan(0-255)')
plt.ylabel('Frekusensi pixel')

max_idx = np.argmax(hist)
plt.axvline(x=max_idx, color='red', linestyle='--', label=f'puncak = {max_idx}')
plt.legend()

plt.show()
```

```python
equalized = cv.equalizeHist(cv.cvtColor(img, cv.COLOR_RGB2GRAY))

plt.figure(figsize=(12,4))

# original image/
plt.subplot(2,2,1)
plt.title('Original')
plt.imshow(img, cmap='gray')

# Equalizec image
plt.subplot(2,2,2)
plt.title('Equalize Image')
plt.imshow(equalized, cmap='gray')

# Sistogram ORIGINAL image
plt.subplot(2,2,3)
plt.hist(img.ravel(), 256, [0, 256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('FrekuensI Pixel')
plt.title('Histogram Asli')

# Histogram IMAGE
plt.subplot(2,2,4)
plt.hist(equalized.ravel(), 256, [0, 256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('FrekuensI Pixel')
plt.title('Tingkat keabuan [0-255]')

plt.show()
```

```python
_, thresh = cv.threshold(equalized, 127, 255, cv.THRESH_BINARY)

plt.figure(figsize=(8,5))

plt.subplot(2,2,1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

plt.subplot(2,2,2)
plt.title('Threshold Image')
plt.imshow(thresh, cmap='gray')

plt.subplot(2,2,3)
plt.title('Original Histogram')
plt.hist(img.ravel(), 256, [0, 256], color='black')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('FrekuensI Pixel')

plt.subplot(2,2,4)
plt.title('Threshold Histogram')
plt.hist(thresh.ravel(), 2, [0, 256], color='black')
plt.xlabel('Tingkat Keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.show()
```

```python
bright = cv.convertScaleAbs(img, alpha=1, beta=50) #cerah
dark = cv.convertScaleAbs(img, alpha=1, beta=50) #gelap

plt.figure(figsize=(10,5))

plt.subplot(2,3,1)
plt.title('Gambar Asli')
plt.imshow(dark, cmap='gray')

plt.subplot(2,3,2)
plt.title('Gambar Cerah')
plt.imshow(bright, cmap='gray')

plt.subplot(2,3,3)
plt.title('Gambar Gelap')
plt.imshow(img, cmap='gray')

plt.subplot(2,3,6)
plt.title('Histogram Gambar Gelap')
plt.hist(dark.ravel(), 256, [0, 256], color='black')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('FrekuensI Pixel')

plt.subplot(2,3,5)
plt.title('Histogram Gambar Terang')
plt.hist(bright.ravel(), 256, [0, 256], color='black')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('FrekuensI Pixel')

plt.subplot(2,3,4)
plt.title('Histogram gambar asli')
plt.hist(img.ravel(), 256, [0, 256], color='black')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('FrekuensI Pixel')

plt.show()
```

```python
inverted = 255 - img #dibalik citra negatif

plt.figure(figsize=(8,5))

plt.subplot(2,2,1)
plt.title('gambar Inverted')
plt.imshow(img, cmap='gray')

plt.subplot(2,2,2)
plt.title("Gambar Inverted")
plt.imshow(inverted, cmap="gray")

plt.subplot(2,2,3)
plt.title('histrogram gambar asli')
plt.hist(img.ravel(), 256, [0,256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.subplot(2,2,4)
plt.title('histrogram gambar negatif')
plt.hist(inverted.ravel(), 256, [0,256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.show()
```

```python
blur = cv.blur(img, (5,5))
edges = cv.Sobel(img, cv.CV_64F, 1,1, ksize=5) #ksize => ganjil

plt.figure(figsize=(10,5))

plt.subplot(2,3,1)
plt.title('Gambar Original')
plt.imshow(img, cmap='gray')

plt.subplot(2,3,2)
plt.title('Gambar Blur')
plt.imshow(blur, cmap='gray')

plt.subplot(2,3,3)
plt.title('Gambar Garis')
plt.imshow(edges, cmap='gray')

plt.subplot(2,3,4)
plt.title('histogram gambar asli')
plt.hist(img.ravel(), 256, [0,256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.subplot(2,3,5)
plt.title('histogram gambar blur')
plt.hist(blur.ravel(), 256, [0, 256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.subplot(2,3,6)
plt.title('histogram gambar edges')
plt.hist(edges.ravel(), 256, [0,256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')
plt.show()
```

```python
norm = cv.normalize(img, None, 0, 255, cv.NORM_MINMAX)
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
plt.title('gambar asli')
plt.imshow(img, cmap='gray')

plt.subplot(2,2,2)
plt.title('gambar normalisasi')
plt.imshow(norm, cmap='gray')

plt.subplot(2,2,3)
plt.title('histogram gambar asli')
plt.hist(img.ravel(), 256, [0, 256], color='red')
plt.xlabel('Tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.subplot(2,2,4)
plt.title('histogram gambar normalisasi')
plt.hist(norm.ravel(), 256, [0, 256], color='red')
plt.xlabel('tingkat keabuan [0-255]')
plt.ylabel('frekuensi pixel')

plt.show
```
